```python
!apt update
!apt install chromium-chromedriver
!pip install selenium

from bs4 import BeautifulSoup
from selenium.webdriver import ActionChains
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd


options = webdriver.ChromeOptions()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# Let's get Income Statement for Microsoft Corporation
index = 'MSFT'

# URL links for Financial Statements
url_is = 'https://finance.yahoo.com/quote/' + index + '/financials?p=' + index
url_bs = 'https://finance.yahoo.com/quote/' + index +'/balance-sheet?p=' + index
url_cf = 'https://finance.yahoo.com/quote/' + index + '/cash-flow?p='+ index
```

```python
# Adding a User Agent to convinve the website that we are not a robot
headers = {
  "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9", 
  "Accept-Encoding": "gzip, deflate", 
  "Accept-Language": "en-GB,en-US;q=0.9,en;q=0.8", 
  "Dnt": "1", 
  "Upgrade-Insecure-Requests": "1", 
  "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_4) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.97 Safari/537.36", 
  "X-Amzn-Trace-Id": "Root=1-5ee7bae0-82260c065baf5ad7f0b3a3e3"
}
# We need to click to Expand All button in order to get all the accounts opened
driver = webdriver.Chrome(options=options)
driver.get(url_is)
button = driver.find_elements_by_xpath("//button[@class='expandPf Fz(s) Bd(0) C($linkColor) C($linkActiveColor):h Fw(500) D(n)--print Fl(end) Mt(5px)']")
for c in button:
  c.click()   
soup_is= BeautifulSoup(driver.page_source,'lxml')
```

```python
soup_is= BeautifulSoup(driver.page_source,'html.parser')
```

```python
ls= [] # Creating an empty list
for s in soup_is.find_all('button'):
  s.extract()
for l in soup_is.find_all(['div']): 
  #Finding all data structure that is div
  ls.append(l.string) # Adding each element one by one to the list

new_ls = list(filter(None,ls))
new_ls = new_ls[20:]
is_data = list(zip(*[iter(new_ls)]*6))
```

```python
is_data
```

```python
Income_st = pd.DataFrame(is_data[0:])
Income_st.columns = Income_st.iloc[0] # Name columns to first row of dataframe
Income_st = Income_st.iloc[1:,] # start to read 1st row
Income_st = Income_st.T # transpose dataframe
Income_st.columns = Income_st.iloc[0] #Name columns to first row of dataframe
Income_st.drop(Income_st.index[0],inplace=True) #Drop first index row
Income_st.index.name = '' # Remove the index name
Income_st.rename(index={'ttm': '6/30/2021'},inplace=True) #Rename ttm in index columns to end of the year
Income_st = Income_st[Income_st.columns[:-5]] # remove last 5 irrelevant columns
Income_st
```

```python
import yfinance as yf

#define the ticker symbol
tickerSymbol = 'MSFT'

#get data on this ticker
tickerData = yf.Ticker(tickerSymbol)

#get the historical prices for this ticker
#One cool feature of yfinance is that you can get highly refined data, all the way down to 5 minute, 3 minute and even 1 minute data!
#1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo
tickerDf = tickerData.history(period='1d', start='2018-08-01', end='2021-08-10')

#see your data
tickerDf.head()
```
